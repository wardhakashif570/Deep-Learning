# The network architecture

```python
from tensorflow.keras import models
from tensorflow.keras import layers
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))#(784 input 512 hidden layer 10 output)
network.add(layers.Dense(10, activation='softmax'))
#Dense layer vector hi leti hai 784 vector and input jo derahai 3 dimensional tensor to model handle ni kr pai ga to hamai use form me daina hai jis me wo accept kre
```

```python
network.compile(optimizer='rmsprop',
loss='categorical_crossentropy',
metrics=['accuracy'])
```

```python
#to bring into 1 dim these steps
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
#one hot encding
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
network.fit(train_images, train_labels, epochs=20, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```
